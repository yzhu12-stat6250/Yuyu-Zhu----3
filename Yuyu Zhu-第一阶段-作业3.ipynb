```python
##Question2: You are given two tables - friend_request and request_accepted. 
##Friend_request contains requester_id, time and sent_to_id and request_accepted table contains time, 
##acceptor_id and requestor_id. How will you determine the overall acceptance rate of requests? 

##Written in Python
import numpy as np
import pandas as pd
file_location = "C:\\Users\\Yuyu Zhu\\Desktop\\"
friend_request_file_name = "friend_request.csv"
request_accepted_file_name = "request_accepted.csv"
friend_request_data = pd.read_csv(file_location + friend_request_file_name)
friend_request_data = np.unique(friend_request_data.loc[:,["requester_id","sent_to_id"]])
request_accepted_data = pd.read_csv(file_location + request_accepted_file_name)
request_accepted_data = np.unique(request_accepted_data.loc[:,["acceptor_id","requestor_id"]])
print(len(request_accepted_data)/len(friend_request_data))
```
